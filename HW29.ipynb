```python
import numpy as np 
import pandas as pd
from IPython.display import display

sa_df = pd.read_csv(r"sport_athletes.csv")
# ------------------------------------------------------
# 1.
display(max(sa_df["Weight(kg)"]))
display(max(sa_df["Height(cm)"]))
display(max(sa_df["Wins"]))
# a, b, c.
sa_df.rename(columns={"Weight(kg)" : "Max_Weight(kg)", "Height(cm)" : "Max_Height(cm)", "Wins" : "Max_Wins"})
# ------------------------------------------------------
# 2.
print(sa_df.groupby('Sport Type')['Weight(kg)'].min().sort_values())
# ------------------------------------------------------
3.
print(sa_df.groupby('Sport Type')['Height(cm)'].nlargest(2))
# ------------------------------------------------------
4.
print(sa_df.groupby(['Sport Type', 'Country']).size())
# a.
display(sa_df.groupby(['Sport Type', 'Country']).max())
# b.
display(sa_df.groupby(['Sport Type', 'Country']).max().loc[("Basketball")])
# c.
display(sa_df.groupby(['Sport Type']).max(numeric_only=True).xs(key='Soccer').reset_index())
# d.
display(sa_df[sa_df['Sport Type'].isin(["Soccer", "Tennis"])].groupby('Sport Type').max(numeric_only=True))
# e.
display(sa_df.groupby(['Sport Type', 'Country']).max(numeric_only=True).swaplevel().sort_index())
# f.
sa_df.groupby('Country').agg({'Wins': ['max', 'min', 'mean', 'std'], 'Height(cm)': ['max', 'min', 'mean', 'std']}).transpose()
# ------------------------------------------------------
# 5.
new_rows = pd.DataFrame({ 'Country': ['Canada', 'USA'], 'Sport Type': ['Hockey', 'Skiing'],
'Athlete Name': ['Wayne Gretzky', 'Lindsey Vonn'], 'Height(cm)': [183, 172],
'Weight(kg)': [84, 68], 'Wins': [3, 4], 'Olympic Date': ['01/02/2002', '12/02/2010'] })
sa_df = pd.concat([sa_df, new_rows], axis=0)
display(sa_df)
# ------------------------------------------------------
# 6.
sa_df['Olympic Date'] = pd.to_datetime(sa_df['Olympic Date'], format='%d/%m/%Y')
display(sa_df.set_index('Olympic Date').resample('YE')['Wins'].sum())
display(sa_df.set_index('Olympic Date').resample('YE')['Wins'].sum().loc[lambda x: x > 0])
display(sa_df.set_index('Olympic Date').resample('YE')['Wins'].sum().replace(0, pd.NA))
# ------------------------------------------------------
# 7.
smoking_df = pd.read_csv(r"athletes_smoking.csv")
new_df = pd.merge(sa_df, smoking_df, on="Athlete Name")
display(new_df)
new_df = pd.merge(sa_df, smoking_df, on="Athlete Name", how="outer")
display(new_df)
```
